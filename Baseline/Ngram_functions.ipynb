```python
import pandas as pd
import glob
"""
The function iterates over the keys (ngram types) in the ngram_files dictionary.
For each ngram type, it iterates over the corresponding file paths from the ngram_files dictionary. 
For each file path, the ngram data is read, and the frequency column name (freq_column_name)
 and ngram column name (ngram_column_name) are set based on the ngram type.
 
The frequency column is initialized with NaN values to ensure all rows have consistent values.

The function then iterates over the "word" column of the dataframe. 
It performs a lookup in the ngram data using ngram_data[ngram_data[ngram_column_name] == word] 
to find the matching row. If a match is found, the frequency value from the corresponding row 
in the "freq" column is assigned to the corresponding row in the frequency column 
(freq_column_name) and the ngram word is assigned to the corresponding row in the ngram column (ngram_column_name).

"""
import pandas as pd
import glob

def add_ngram_frequencies(dataframe):
    ngram_files = {
        1: glob.glob("corpus/eng_Ngram_corpus/1grams/*.csv"),
        2: glob.glob("corpus/eng_Ngram_corpus/2grams/*.csv"),
        3: glob.glob("corpus/eng_Ngram_corpus/3grams/*.csv"),
        4: glob.glob("corpus/eng_Ngram_corpus/4grams/*.csv"),
        5: glob.glob("corpus/eng_Ngram_corpus/5grams/*.csv")
    }
    
    for n in range(1, 6):
        for file_path in ngram_files[n]:
            ngram_data = pd.read_csv(file_path)
            freq_column_name = f"{n}gram_freq"
            ngram_column_name = f"{n}gram"
            
            # Extract the filename (ngram name) from the file path
            ngram_name = file_path.split("/")[-1].split(".")[0]
            
            # Initialize the frequency column with NaN values
            dataframe[freq_column_name] = float('nan')
            
            # Iterate over the "word" column
            for index, row in dataframe.iterrows():
                word = row['word']
                
                # Look up the frequency in the ngram data
                matching_row = ngram_data[ngram_data['ngram'] == word]
                if not matching_row.empty:
                    frequency = matching_row.iloc[0]['freq']
                    dataframe.at[index, freq_column_name] = frequency
                    dataframe.at[index, ngram_column_name] = word
    
    return dataframe
```

```python
from textblob import TextBlob
from textblob.exceptions import NotTranslated

def assign_pos_tags(text):
    try:
        blob = TextBlob(text)
        blob = blob.translate(to='en')  # Translate text to English if needed
        pos_tags = blob.tags
        return pos_tags
    except NotTranslated:
        print("Unable to translate the text to English.")
    except Exception as e:
        print(f"An error occurred: {str(e)}")

# Example usage for a 3-gram
trigram = "example of text"
pos_tags = assign_pos_tags(trigram)
print(pos_tags)

# Example usage for a 4-gram
fourgram = "this is an example"
pos_tags = assign_pos_tags(fourgram)
print(pos_tags)

# Example usage for a 5-gram
fivegram = "this is an example sentence"
pos_tags = assign_pos_tags(fivegram)
print(pos_tags)
```

```python
import spacy

def assign_pos_tags(text):
    nlp = spacy.load('en_core_web_sm')
    doc = nlp(text)
    pos_tags = [(token.text, token.pos_) for token in doc]
    return pos_tags

# Example usage for a 3-gram
trigram = "example of text"
pos_tags = assign_pos_tags(trigram)
print(pos_tags)

# Example usage for a 4-gram
fourgram = "this is an example"
pos_tags = assign_pos_tags(fourgram)
print(pos_tags)

# Example usage for a 5-gram
fivegram = "this is an example sentence"
pos_tags = assign_pos_tags(fivegram)
print(pos_tags)
```
