```python
import pandas as pd
import nltk
from nltk.corpus import stopwords


# Get English stop words
stop_words = set(stopwords.words('english'))



```

```python

output = []

for i, row in enumerate(MWEs['split']):
    row_dict = {}
    for word in row:
        # Convert word to lowercase
        word = word.lower()
        # Check if word contains a hyphen
        if '-' in word:
            # Split word and create a key for each part
            for part in word.split('-'):
                # Only add part to dict if it's not a stop word
                if part not in stop_words:
                    # Check if part is in 'phrase' column of word_parse_features DataFrame
                    if part in word_parse_features['phrase'].values:
                        # Get row from word_parse_features DataFrame and convert it to dictionary
                        row_dict[part] = word_parse_features[word_parse_features['phrase'] == part].to_dict('records')[0]
                    else:
                        # If part not found, add it with empty dictionary value
                        row_dict[part] = {}
        else:
            # Only add word to dict if it's not a stop word
            if word not in stop_words:
                # Check if word is in 'phrase' column of word_parse_features DataFrame
                if word in word_parse_features['phrase'].values:
                    # Get row from word_parse_features DataFrame and convert it to dictionary
                    row_dict[word] = word_parse_features[word_parse_features['phrase'] == word].to_dict('records')[0]
                else:
                    # If word not found, add it with empty dictionary value
                    row_dict[word] = {}
    output.append(row_dict)
    # Check if the dictionary is empty
    if not any(row_dict.values()):
        print(f"Dictionary for row {i+1} is empty: {row_dict}")

# Add the output list as a new column to the MWEs DataFrame
MWEs['MWE_feats'] = output
```
