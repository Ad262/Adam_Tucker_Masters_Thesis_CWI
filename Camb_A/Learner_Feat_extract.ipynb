```python
# This code uses dicts and batches to optimise performance. 
"""

function to save each batch of the DataFrame to a separate temporary pickle file,
 and then concatenate all the temporary pickle files at the end to get the final output file. 
 This outputs _temp.pkl and then concatinates them. 
 
"""
import os
import pandas as pd
import re

def count_target_occurrences(input_folder, output_folder, batch_size=1000):
    # Create the output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    # Read the "filtered_Lang_8_data.csv" file
    csv_file_path = os.path.join(input_folder, "filtered_Lang_8_data.csv")
    df_csv = pd.read_csv(csv_file_path)

    # Initialize the target counts dictionary
    target_counts = {}

    # Generator function to process the DataFrame in batches
    def batch_generator(df, batch_size):
        num_batches = (len(df) // batch_size) + 1
        for i in range(num_batches):
            yield df.iloc[i * batch_size: (i + 1) * batch_size]

    # Iterate over .pkl files in the input folder
    for file_name in os.listdir(input_folder):
        if file_name.endswith('.pkl'):
            # Construct the file paths
            input_file_path = os.path.join(input_folder, file_name)

            # Load the DataFrame from the input file
            df_pkl = pd.read_pickle(input_file_path)

            # Process the DataFrame in batches and update the target counts dictionary
            for batch_df in batch_generator(df_pkl, batch_size):
                for phrase in batch_df["phrase"]:
                    target = str(phrase)  # Use the target word from the "phrase" column
                    if target in target_counts:
                        target_counts[target] += df_csv.iloc[:, 4].str.count(re.escape(target)).sum()
                    else:
                        target_counts[target] = df_csv.iloc[:, 4].str.count(re.escape(target)).sum()

                # Add the "Lang-8_freq" column to the batch DataFrame based on the target counts
                batch_df["Lang-8_freq"] = batch_df["phrase"].apply(lambda target: target_counts.get(str(target), 0))

                # Save the updated batch DataFrame to a temporary pickle file
                temp_file_path = os.path.join(output_folder, f"{file_name}_temp.pkl")
                batch_df.to_pickle(temp_file_path)

    # Concatenate all the temporary pickle files to get the final output file
    for file_name in os.listdir(output_folder):
        if file_name.endswith('_temp.pkl'):
            temp_file_path = os.path.join(output_folder, file_name)
            df_temp = pd.read_pickle(temp_file_path)

            # Save the temporary DataFrame to the final output file
            output_file_path = os.path.join(output_folder, file_name.replace('_temp.pkl', '.pkl'))
            df_temp.to_pickle(output_file_path)

            # Remove the temporary pickle file
            os.remove(temp_file_path)
```

```python
import os
import pandas as pd

def add_frequency_column(folder_path, phrase_column_name='phrase', text_corrected_column_name='text_corrected'):
    # Get the list of .pkl files in the folder
    pkl_files = [file for file in os.listdir(folder_path) if file.endswith('.pkl')]

    # Load the data from "Corpus/EFCAMDAT_A1.pkl" to get the phrase frequency
    path_to_corpus = os.path.join(folder_path, 'Corpus/EFCAMDAT_A1.pkl')
    corpus_data = pd.read_pickle(path_to_corpus)

    # Group by the "text_corrected" column and count the occurrences of each phrase
    phrase_frequency = corpus_data.groupby(text_corrected_column_name).size().reset_index(name='A1_freq')

    # Iterate through each .pkl file and add the frequency column
    for pkl_file in pkl_files:
        file_path = os.path.join(folder_path, pkl_file)
        df = pd.read_pickle(file_path)
        df = df.merge(phrase_frequency, on=text_corrected_column_name, how='left')
        df.rename(columns={'A1_freq': 'A1_freq'}, inplace=True)

        # Save the modified DataFrame back to the .pkl file
        df.to_pickle(file_path)

if __name__ == "__main__":
    folder_path = "Corpus/All_features_pp"
    add_frequency_column(folder_path)
```

In this updated version, we count how many times each word from the "phrase" column occurs in the "text_corrected" column of the "Corpus/EFCAMDAT_A1.pkl" file. Then, we add the corresponding count to the "A1_freq" column in the original .pkl files in the "Lang8_feat/" folder. The new files will be saved with "_EFCAMDAT" added to their filenames.

```python


# Work functions to return all CEFR data from EFCAMDAT. Adds 5 new columns to existing 38 columns of CAMB features.  

import os
import pandas as pd

def get_cefr_freq_A1(folder_path, phrase_column_name='phrase', text_corrected_column_name='text_corrected'):
    return get_cefr_freq(folder_path, phrase_column_name, text_corrected_column_name, 'EFCAMDAT_A1.pkl', 'A1_freq')

def get_cefr_freq_A2(folder_path, phrase_column_name='phrase', text_corrected_column_name='text_corrected'):
    return get_cefr_freq(folder_path, phrase_column_name, text_corrected_column_name, 'EFCAMDAT_A2.pkl', 'A2_freq')

def get_cefr_freq_B1(folder_path, phrase_column_name='phrase', text_corrected_column_name='text_corrected'):
    return get_cefr_freq(folder_path, phrase_column_name, text_corrected_column_name, 'EFCAMDAT_B1.pkl', 'B1_freq')

def get_cefr_freq_B2(folder_path, phrase_column_name='phrase', text_corrected_column_name='text_corrected'):
    return get_cefr_freq(folder_path, phrase_column_name, text_corrected_column_name, 'EFCAMDAT_B2.pkl', 'B2_freq')

def get_cefr_freq_C1(folder_path, phrase_column_name='phrase', text_corrected_column_name='text_corrected'):
    return get_cefr_freq(folder_path, phrase_column_name, text_corrected_column_name, 'EFCAMDAT_C1.pkl', 'C1_freq')

def get_cefr_freq(folder_path, phrase_column_name, text_corrected_column_name, cefr_file, cefr_freq_column):
    # Load the data from the specified cefr_file to get the phrase frequency
    path_to_corpus = os.path.join(folder_path, '../Corpus', cefr_file)
    corpus_data = pd.read_pickle(path_to_corpus)

    # Create a dictionary to store the word frequencies from the "text_corrected" column
    word_freq_dict = dict(corpus_data[text_corrected_column_name].str.split(expand=True).stack().value_counts())

    # Iterate through each .pkl file and add the frequency column
    for pkl_file in os.listdir(folder_path):
        if pkl_file.endswith('.pkl') and pkl_file != cefr_file:
            file_path = os.path.join(folder_path, pkl_file)
            df = pd.read_pickle(file_path)

            # Count the occurrences of each word from the "phrase" column in the "text_corrected" column
            df[cefr_freq_column] = df[phrase_column_name].apply(lambda x: sum(word_freq_dict.get(word, 0) for word in x.split()))

            # Save the modified DataFrame to the same .pkl file
            df.to_pickle(file_path)

            # Save the modified DataFrame to a corresponding .csv file
            csv_file_path = os.path.splitext(file_path)[0] + '_cefr.csv'
            df.to_csv(csv_file_path, index=False)

    return None  # No need to return merged_df anymore

if __name__ == "__main__":
    folder_path = "Lang8_feat_test/"
    # Call all functions to extract data for different CEFR levels
    get_cefr_freq_A1(folder_path)
    get_cefr_freq_A2(folder_path)
    get_cefr_freq_B1(folder_path)
    get_cefr_freq_B2(folder_path)
    get_cefr_freq_C1(folder_path)
```
